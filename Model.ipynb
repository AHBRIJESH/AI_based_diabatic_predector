# Import Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
```

# Import the Data to the Environment

```python
data = pd.read_csv('diabetes.csv')
data.head()
```

```python
X = data.drop('Outcome',axis='columns')
X.head()
```

```python
target= data.Outcome
target.head()
```

# Creting the Model

```python
clf = tree.DecisionTreeClassifier()
clf.fit(X,target)
```

```python
clf.score(X,target)
```

```python
a = int(input("Number of Pregnancies: "))
b = int(input("Glucose Level: "))
c = int(input("Blood Pressure: "))
d = int(input("Skin Thickness: "))
e = int(input("Insulin Level: "))
f = float(input("Body Mass Index: "))
g = float(input("Diabetic Pedigree Function: "))  
h = int(input("Age: "))

if(clf.predict([[a,b,c,d,e,f,g,h]])==1):
    print("The patient has the higher chance of getting Diabetes")
else:
    print("The patient has the Lower chance of getting Diabetes")
```
